```python
import pandas as pd
from transformers import pipeline
```

```python
## 
dat = pd.read_csv("../results/data_cleaned.txt", delimiter = "\t")
dat
```

```python
company_revs = dat["company_review"].to_list()
```

```python
sentiment_classifier = pipeline(
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student", 
    return_all_scores=True)
```

```python
final_res = []

for index, row in dat.iterrows():

    try:
        i_res = sentiment_classifier(row["company_review"])
        i_pos_score = i_res[0][0]["score"]
        i_neu_score = i_res[0][1]["score"]
        i_neg_score = i_res[0][2]["score"]
        i_final_res = {'id': row["id"], 'positive': i_pos_score, 'nutral': i_neu_score, 'negative': i_neg_score}
        final_res.append(i_final_res)
        
    except Exception as e:
        # Handle the exception (e.g., log it, print a message)
        print(f"Error processing : {e}")
```

```python
final_res_df = pd.DataFrame(final_res)
```
