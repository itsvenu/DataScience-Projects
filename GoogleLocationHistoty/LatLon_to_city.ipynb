```python
## Long, Lat GPS coordinates to city name
## https://github.com/thampiman/reverse-geocoder
## Awesome tool, works offline

import pandas as pd
import reverse_geocoder as rg

## read lat, lon as pandas df

df  = pd.read_csv("glh_clean_v01_LatLon_table.txt", sep="\t", header=0)

## look at head of data

df.head()
```

```python
## Convert above dataframe to tuple, e.g. (49.401803, 8.680858), (49.401790,8.680845) ...

res = [tuple(x) for x in df.values]
res_tuple = tuple(res)

## search for city names
city_results = rg.search(res_tuple)

city_results_df = pd.DataFrame(city_results)
```

```python
city_results_df.head()
```

```python
city_results_df.tail()
```

```python
## add initial dataframe to results as column-bind
## Why? results only show closest major city coordinates
## I need exact coordinates to add to my results

final_res = pd.concat([df.reset_index(drop=True), city_results_df], axis=1)

final_res.head()
```

```python
final_res.tail()
```

```python
final_res.to_csv("glh_clean_v01_LatLon_table_citynames.txt", encoding='utf-8', index=False, sep = "\t")
```
